```python
#example 1: simple code
from kafka import KafkaProducer, KafkaConsumer

# Create a Kafka producer
producer = KafkaProducer(bootstrap_servers='localhost:9092')

# Publish messages to a Kafka topic
topic = 'my_topic'
messages = ['message1', 'message2', 'message3']

for message in messages:
    producer.send(topic, message.encode('utf-8'))
    print(f"Sent message: {message}")

# Create a Kafka consumer
consumer = KafkaConsumer(topic, bootstrap_servers='localhost:9092', group_id='my_group')

# Consume messages from the Kafka topic
for message in consumer:
    print(f"Received message: {message.value.decode('utf-8')}")
```

```python
## example2: data of csv file
from kafka import KafkaProducer, KafkaConsumer
import csv
import json

# Create a Kafka producer
producer = KafkaProducer(bootstrap_servers='localhost:9092', value_serializer=lambda v: json.dumps(v).encode('utf-8'))

# Create a Kafka consumer
consumer = KafkaConsumer('my_topic', bootstrap_servers='localhost:9092', group_id='my_group',
                         value_deserializer=lambda m: json.loads(m.decode('utf-8')))

# Open the CSV file
csv_file_path = './fraud_data.csv'

with open(csv_file_path, 'r') as file:
    # Create a CSV reader
    reader = csv.reader(file)

    # Skip the header row if needed
    next(reader)

    # Iterate over the rows in the CSV file
    for row in reader:
        # Extract data from each column of the row
        column1 = row[0]
        column2 = row[1]
        # ...

        # Create a dictionary or any desired structure to represent the extracted data
        data = {
            'column1': column1,
            'column2': column2,
            # ...
        }

        # Publish the data to the Kafka topic
        topic = 'my_topic'
        producer.send(topic, value=data)
        print(f"Published data to Kafka: {data}")

# Wait for all messages to be sent before proceeding
producer.flush()

# Consume messages from the Kafka topic
for message in consumer:
    data = message.value
    print(f"Received data from Kafka: {data}")
```

```python
#example3: for data extraction from website
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate
from kafka import KafkaProducer, KafkaConsumer

# Create a Kafka producer
producer = KafkaProducer(bootstrap_servers='localhost:9092', value_serializer=lambda v: v.encode('utf-8'))

# Create a Kafka consumer
consumer = KafkaConsumer('my_topic', bootstrap_servers='localhost:9092', group_id='my_group',
                         value_deserializer=lambda m: m.decode('utf-8'))

# URL of the website to extract data from
url = 'https://www.france24.com/'

# Phrase to filter and extract from the website
phrase = 'artificial intelligence'

# Send HTTP GET request to the website
response = requests.get(url)
html_content = response.text

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find all the links containing the phrase
links_with_phrase = [(link.text, link['href']) for link in soup.find_all('a', href=True) if phrase in link.text]

# Print the links containing the phrase
print("Links containing the phrase 'artificial intelligence':")
for index, (link_text, link_url) in enumerate(links_with_phrase, start=1):
    print(f"{index}. {link_text}: {link_url}")

# Prepare the data as a table
table_data = []
for index, (link_text, link_url) in enumerate(links_with_phrase, start=1):
    table_data.append([index, link_text, link_url])

table = tabulate(table_data, headers=['Index', 'Link Text', 'Link URL'], tablefmt='psql')

# Publish the table data to the Kafka topic
topic = 'my_topic'
producer.send(topic, value=table)
print("Published table data to Kafka")

# Wait for all messages to be sent before proceeding
producer.flush()

# Consume messages from the Kafka topic
for message in consumer:
    data = message.value
    print(f"Received data from Kafka:\n{data}")

    # Process the consumed data or perform any desired operations
    # ...

# Close the Kafka consumer
consumer.close()
```
